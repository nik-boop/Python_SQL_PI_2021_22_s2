````python
from sqlalchemy import MetaData, Table, create_engine, select, and_, or_, not_, text
from sqlalchemy.sql import func, distinct, exists
import pandas


# Сброс ограничений на количество выводимых рядов
pandas.set_option('display.max_rows', None)
 
# Сброс ограничений на число столбцов
pandas.set_option('display.max_columns', None)
 
# Сброс ограничений на количество символов в записи
pandas.set_option('display.max_colwidth', None)
````

````python
engine = create_engine('sqlite:///../Data/Students.db')
metadata_obj = MetaData() # Объект-контейнер, содержит информацию о схеме данных
metadata_obj.reflect(bind=engine) # Отображение БД
````

Создание объектов таблиц

````python
exam_marks=Table("exam_marks", metadata_obj, autoload=True, autoload_with="engine")
lecturer=Table("lecturer", metadata_obj, autoload=True, autoload_with="engine")
student=Table("student", metadata_obj, autoload=True, autoload_with="engine")
subj_lect=Table("subj_lect", metadata_obj, autoload=True, autoload_with="engine")
subject=Table("subject", metadata_obj, autoload=True, autoload_with="engine")
university=Table("university", metadata_obj, autolad=True, autolad_with="engine")

st = student.alias()
em = exam_marks.alias()
su = subject.alias()
un = university.alias()
le = lecturer.alias()
sl = subj_lect.alias()
````

````python
conn=engine.connect() # Соединение
````

# Решение

````python
# 1. Напишите запрос, который выполняет вывод фамилий студентов,
# сдававших экзамены, вместе с идентификаторами каждого сданного ими предмета обучения.
````

````python
# 2. Напишите запрос, который выполняет выборку фамилий всех студентов,
# с указанием для студентов, сдававших экзамены, идентификаторов сданных ими предметов обучения.
s = select(st.c.surname, em.c.subj_id)
s = s.join_from(st, em, isouter=True)
s = s.order_by(st.c.surname)

print(s)
resoult = conn.execute(s).fetchall()
print(f"\nCount answer: {len(resoult)}")
pandas.read_sql(s, conn)
````

````python
# 3. Напишите запрос, который выполняет вывод фамилий студентов,
# сдававших экзамены, вместе с наименованиями каждого сданного ими предмета обучения.
s = select(st.c.surname, su.c.subj_id)
s = s.join(em, em.c.student_id==st.c.student_id)
s = s.join(su)
s = s.where(em.c.mark>2)
s = s.order_by(st.c.surname)

print(s)
resoult = conn.execute(s).fetchall()
print(f"\nCount answer: {len(resoult)}")
pandas.read_sql(s, conn)
````

````python
# 4. Напишите запрос на выдачу списка всех студентов. Для студентов,
# сдававших экзамены, укажите названия соответствующих предметов обучения.

s = select(st.c.surname, em.c.subj_id, su.c.subj_name)
s = s.outerjoin_from(em, st, em.c.student_id==st.c.student_id)
s = s.outerjoin(su)
s = s.order_by(st.c.surname)

print(s)
resoult = conn.execute(s).fetchall()
print(f"\nCount answer: {len(resoult)}")
pandas.read_sql(s, conn)
````

````python
# 5. Напишите запрос на выдачу названий всех предметов,
# по которым студенты получили только хорошие (4 и 5) оценки.
# В выходных данных должны быть приведены фамилии студентов, названия предметов и оценка.

sq = select(func.max(em.c.mark))
sq = sq.where(em.c.student_id==st.c.student_id)

s = select(st.c.surname, su.c.subj_name, em.c.mark)
s = s.where(em.c.mark>3)
s = s.join(su)
s = s.join(st)
s = s.order_by(st.c.surname)

print(s)
resoult = conn.execute(s).fetchall()
print(f"\nCount answer: {len(resoult)}")
pandas.read_sql(s, conn)
````

````python
# 6. Напишите запрос, который выполняет вывод списка университетов с рейтингом,
# превышающим 300, вместе со значением максимального размера стипендии, получаемой студентами в этих университетах.
````

````python
# 7. Напишите запрос на выдачу списка студентов (в алфавитном порядке фамилий)
# вместе со значением рейтинга университета, где каждый из них учится,
# включив в список и тех студентов, место учебы которых в базе данных не указано.

s = select(st.c.surname, un.c.univ_name, un.c.rating)
s = s.outerjoin(un, st.c.univ_id==un.c.univ_id)
s = s.order_by(st.c.surname)

print(s)
resoult = conn.execute(s).fetchall()
print(f"\nCount answer: {len(resoult)}")
pandas.read_sql(s, conn)

````

````python
# 8. Напишите запрос для получения списка всех студентов вместе с названиями университетов,
# в которых они учатся. Отдельным запросом получите записи,
# расширяющие данный список по сравнению с тем, который был бы получен внутренним соединением.

s = select(st.c.surname, un.c.univ_name)
s = s.join(un, st.c.univ_id==un.c.univ_id)
s = s.order_by(st.c.surname)

print(s)
resoult = conn.execute(s).fetchall()
print(f"\nCount answer: {len(resoult)}")
display(pandas.read_sql(s, conn))

st2 = student.alias()
un2 = university.alias()

sq = select(st2.c.student_id)
sq = sq.join(un2, st.c.univ_id==un2.c.univ_id)
sq = sq.where(st2.c.student_id==st.c.student_id).exists()

s = select(st.c.surname, un.c.univ_name)
s = s.outerjoin(un, st.c.univ_id==un.c.univ_id)
s = s.where(~sq)
s = s.order_by(st.c.surname)

print(s)
resoult = conn.execute(s).fetchall()
print(f"\nCount answer: {len(resoult)}")
pandas.read_sql(s, conn)
````

````python
# 9. Напишите запрос для получения списка всех университетов вместе с фамилиями студентов, в них обучающихся.
# Отдельным запросом получите записи, расширяющие данный список по сравнению с тем,
# который был бы получен внутренним соединением.

s = select(st.c.surname, un.c.univ_name)
s = s.join_from(un, st, st.c.univ_id==un.c.univ_id)
s = s.order_by(st.c.surname)

print(s)
resoult = conn.execute(s).fetchall()
print(f"\nCount answer: {len(resoult)}")
display(pandas.read_sql(s, conn))


st2 = student.alias()
un2 = university.alias()

sq = select(un2.c.univ_id)
sq = sq.join_from(un2, st2)
sq = sq.where(un2.c.univ_id==un.c.univ_id).exists()

s = select(st.c.surname, un.c.univ_name)
s = s.outerjoin_from(un, st, st.c.univ_id==un.c.univ_id)
s = s.where(~sq)
s = s.order_by(st.c.surname)

print(s)
resoult = conn.execute(s).fetchall()
print(f"\nCount answer: {len(resoult)}")
pandas.read_sql(s, conn)
````
