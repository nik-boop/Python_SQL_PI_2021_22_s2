````python
# 5.1. Покажите длительность с накопительным итогом по поездкам, начавшимся в интервале с 10:00 до 11:30,
# в порядке возрастания времени начала поездки.

s3 = '''

SELECT SUM(duration) OVER ( 
    ORDER BY started_at 
    ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) as accumulative_res, BikeShare.*
FROM BikeShare
WHERE strftime("%H:%M", started_at) BETWEEN "10:00" AND "11:29"

'''


with engine.connect() as conn:
    result = conn.execute(text(s3))
    display(pd.read_sql(text(s3), conn))
````

````python

# 5.2. Ранжируйте пары станций отправления и назначения по количеству поездок.
# Паре с большим количеством поездок присвойте меньший номер ранга.

s3 = '''

With A as (
SELECT count(*) as count, BikeShare.start_station_id, BikeShare.end_station_id, BikeShare.start_station_name, BikeShare.end_station_name
from BikeShare
GROUP by BikeShare.start_station_id, BikeShare.end_station_id
)

SELECT dense_rank() over(Order by A.count desc) as ranc, A.*
FROM A;

'''


with engine.connect() as conn:
    result = conn.execute(text(s3))
    display(pd.read_sql(text(s3), conn))
````

````python
# 5.3. Покажите, сколько в проценках занимает время каждой поездки от суммарного времени,
# затраченного на все поездки между станциями данной пары.

s3 = '''

SELECT ride_id, start_station_id, end_station_id,
CAST(duration*100/sum(duration) OVER (PARTITION BY start_station_id, end_station_id) AS VARCHAR2(2)) || "%" as perc_time
FROM BikeShare;


'''


with engine.connect() as conn:
    result = conn.execute(text(s3))
    display(pd.read_sql(text(s3), conn))
````

````python
# 5.4. Покажите 4 самых долгих поездки с каждой станции отправления,
# упорядоченные по убыванию длительности поездки для данной станции.

s3 = '''

SELECT *
FROM (
    SELECT BikeShare.ride_id, BikeShare.start_station_id, duration, row_number() OVER (PARTITION BY start_station_id order by duration desc) as pr
    FROM BikeShare
    ORDER BY duration desc) as t1
WHERE pr <= 4
GROUP BY t1.start_station_id, t1.pr;

'''


with engine.connect() as conn:
    result = conn.execute(text(s3))
    display(pd.read_sql(text(s3), conn))
````

````python
# 5.5. Разбейте поездки на квантили по длительности, упорядочив по возрастанию длительности.
# Распечатайте следующие поля результата: номер квантиля, идентификатор поездки, станция назначения, длительность.

s3 = '''

with t1 as (
    SELECT ride_id, end_station_name, duration
    FROM bikeshare
    ORDER BY duration desc
),

t2 as (
    SELECT NTILE (4) OVER (ORDER BY t1.duration) as RowNum, t1.*
    FROM t1
)

SELECT t2.*
FROM t2;

'''


with engine.connect() as conn:
    result = conn.execute(text(s3))
    display(pd.read_sql(text(s3), conn))
````
