#### Задание № 9.
Дана статистика расходов на рекламную кампанию по каналу ('source'), расходам ('cost') и прибыли от рекламы ('profit').
```
ads = [
    {'cost':98, 'source':'vk', 'profit':150},
    {'cost':153, 'source':'yandex', 'profit':270},
    {'cost':110, 'source':'facebook', 'profit':201}
]
```
Найти канал, обеспечивший максимальную эффективность = прибыль/расходы. Распечатать название канала и значение эффективности, округленное до двух десятичных знаков.

````python
ads = [
    {'cost':98, 'source':'vk', 'profit':150},
    {'cost':153, 'source':'yandex', 'profit':270},
    {'cost':110, 'source':'facebook', 'profit':201}
]

# Ваш код здесь

max_eff, name_ch = 0, ""
for dt in ads:
    if ((dt['profit']) / (dt['cost'])) > max_eff:
        max_eff = round((dt['profit']) / (dt['cost']), 2)
        name_ch = dt['source']
print(f"Максимальную эффективность в {max_eff} показал канал {name_ch}.")
````

#### Задание № 10.
Представить данные Задачи № 9 в табличном виде, вывести эффективность в отдельный столбец и найти максимально эффективную организацию, оперируя данными таблицы.

````python
import pandas as pd
````

````python
ads = [
    {'cost':98, 'source':'vk', 'profit':150},
    {'cost':153, 'source':'yandex', 'profit':270},
    {'cost':110, 'source':'facebook', 'profit':201}
]

main_dict = {"cost": [], "source": [], "profit": [], "effect": []}
for dt in ads:
    main_dict["cost"].append(dt["cost"])
    main_dict["source"].append(dt["source"])
    main_dict["profit"].append(dt["profit"])
    main_dict["effect"].append(round((dt['profit']) / (dt['cost']), 2))
    
main = pd.DataFrame(main_dict)
main
````

````python
print(f"Максимальная эффективность - {max(main_dict['effect'])}.")
````

# Домашняя работа.
Сгруппировать все здания библиотек по административному округу Москвы (с учётом того, что несколько библиотек может быть в одном здании по совершенно одинаковому адресу и индексу).

````python
import json
import pandas as pd
````

````python
with open('./Data/Libraries.json') as f:
    libraries = json.load(f)
````

````python
s_adm, dic_adm = set(), {}

# Составим список всех округов Москвы при помощи фильтрации дубликатов через множество

for i in range(len(libraries) - 1):
    elem = [libraries[i]["ObjectAddress"][0]["AdmArea"]]
    s_adm.update(elem)

s_adm = [i.split()[0] for i in s_adm]

# Готовим итоговый вариант группировки библиотек по округам через словарь

for i in s_adm:
    dic_adm[i] = []
    
print(dic_adm)
````

````python
# Группируем все наши библиотеки в Москве по округу
# и составляем итоговую структуру в словарь, значения которого - списки библиотек округа с конкретными адресами в городе

for i in range(len(libraries) - 1):
    elem = libraries[i]["ObjectAddress"][0]["AdmArea"].split()[0]
    if elem in s_adm:
        dic_adm[elem].append(libraries[i]["CommonName"] + ", Адрес: " + libraries[i]["OrgInfo"][0]["LegalAddress"])
````

````python
# Устанавливаем максимальную длину получившихся списков библиотек

len_lst = []
for i in s_adm:
    len_lst.append(len(dic_adm[i]))
max_len = max(len_lst)

# Далее сверимся, какая максимальная длина вышла и какая длина у остальных элементов списка
````

````python
# Дополняем значениями Nan недостающие элементы в остальных списках, чтобы уравнять длину 

for i in s_adm:
    counter = len(dic_adm[i])
    while counter != max_len:
        dic_adm[i].append("NaN")
        counter += 1
````

````python
# Сверяемся последний раз с получившимися результатами

len_lst_final = []
for i in s_adm:
    len_lst_final.append(len(dic_adm[i]))
````
