<p style="color:#000000">
    Имеется набор данных "Библиотеки", в котором в формате JSON содержатся сведения о библиотеках Москвы. Одна библиотека может располагаться в нескольких зданиях.
    
Определить, в каких полях JSON-файла содержатся сведения об админстративном округе, в котором находится конкретное здание библиотеки.
    
Сгруппировать здания библиотек по административному округу и определить, сколько зданий находится в каждом округе.
</p>

```python
import ast
import pandas

pandas.set_option('display.max_rows', None)
pandas.set_option('display.max_columns', None)
pandas.set_option('display.width', None)

f = open('../Data/data-7361-2022-01-19.json')
data = f.read()
f.close()
data = ast.literal_eval(data) 


dict_admarea = {}
DF = pandas.DataFrame(data)
DF.loc[:, ['FullName', 'ObjectAddress']].apply(
    lambda x, dct: [dct.setdefault(Object['AdmArea'], []).append([Object['Address'], x['FullName']]) for Object in x['ObjectAddress']],
    args=(dict_admarea, ),
    axis=1
)

d_f = pandas.DataFrame(columns=['AdmArea', 'Address', 'FullName'])
index = 0
for key in dict_admarea:
    # print(*dict_admarea[key], sep='\n')
    for adr in dict_admarea[key]:
        d_f.loc[index, 'AdmArea'] = key
        d_f.loc[index, 'Address'] = adr[0]
        d_f.loc[index, 'FullName'] = adr[1]
        
        index+=1
display(d_f.groupby('AdmArea')['Address'].describe()['count'])

# Пример БД по Центральному администронивному округу
display(d_f.loc[d_f['AdmArea'] == 'Центральный административный округ'].sort_values('Address'))
# БД
# display(d_f)
```

<p style="font-size:1.6em; color:#FF4E00"> Задание 2 </p>

Продолжим работу с набором данных "Библиотеки". Как мы уже выяснили, одна библиотека может располагаться более, чем в одном здании.

Найти все здания, ПОЛНОСТЬЮ оборудованные тактильными средствами информации для инвалидов по зрению. Распечатать адреса всех таких зданий с названиями библиотек.

```python
import pandas
import ast

pandas.set_option('display.max_rows', None)
pandas.set_option('display.max_columns', None)
pandas.set_option('display.width', None)

f = open('../Data/data-7361-2022-01-19.json')
data = f.read()
f.close()
data = ast.literal_eval(data) 

dict_admarea = {}
DF = pandas.DataFrame(data)
DF.loc[:, ['FullName', 'ObjectAddress']].apply(
    lambda x, dct: [dct.setdefault(Object['Address'], []).append(x['FullName']) for Object in x['ObjectAddress'] if Object['Availability'][0]['available_z'] == 'полностью'],
    args=(dict_admarea, ),
    axis=1
)

d_f = pandas.DataFrame(columns=['Address', 'FullName'])
index = 0
for key in dict_admarea:
    for adr in dict_admarea[key]:
        d_f.loc[index, 'Address'] = key
        d_f.loc[index, 'FullName'] = adr
        
        index+=1
display(d_f.groupby('Address').describe())
```
